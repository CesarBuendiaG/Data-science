El problema (potencial) de una compañia al querer recomendar juegos a personas que llegan a su plataforma
- Para nuevos usuarios, seguerir nuevos juegos (solo lo que pueda interesarle) al ingresar a la plataforma solo 1 vez al mes
- No sabemos el comportamiento del usuario (al ser nuevo) por lo que no sabemos que prefiere o no comprar
- Generar una prediccion dependiendo de algunos datos del usuario
- Solo ML


Cuento con el archivo base (data limpia) para esta practica con datos de algunos usuarios:
- Edad
- Genero (0=hombre / 1=mujer)
- Juegos (accion,fps,jrpg,plataforma,puzzle)

```python
import pandas as pd
df = pd.read_csv("fuentes/juegos.csv")
```

```python
df
```

Al ser datos limpios podemos darnos cuenta de algunas cosas:
- hombres entre 8-11 y 31-35 años prefieren juegos de accion
- hombres entre 12 y 18 años prefieren juegos de fps
- hombres entre 22 y 28 años prefieren juegos de jrpg
- no hay registro de hombres que prefieran juegos de plataforma o puzzle

- mujeres entre 10 y 14 años prefieren juegos de plataforma
- mujeres entre 15 y 18 años prefieren juegos de puzzle
- mujeres entre 19 y 25 años prefieren juegos de fps
- no hay registro de mujeres que prefieran juegos de accion o jrpg


El siguiente paso seria limpiar los datos (nulos, duplicados, incorrectos, no relevantes, etc.), sin embargo en esta practica asumiremos que estos datos ya se encuentran limpios

En cambio, procederemos a divir los datos en:
- Datos de entrada (lo que le voy a pasar al modelo)
- Datos de salida (lo que el modelo me va a responder)

En este caso, mis datos de entrada van a ser "edad" y "genero", y los datos de salida seran "juegos"
- Por ejemplo, yo le quiero pasar la edad de 8 y genero 0, lo que el modelo debe de regresar son juegos accion


```python
# Tal y como se ve acá
df.head(1)
```

```python
# Para eso uso el metodo .drop, lo que hace es crear una tabla, por lo que hay que asignarselo a una nueva variable
# *Por convencion, la variable de los imputs se llama "X"
X = df.drop(columns=["juegos"])
X
```

```python
# Ahora procedo a generar mi tabla de datos de salida, como solo es una columna, puedo acceder directamente a ella
# *Por convencion la tabla de salidas se llama "y"
y = df["juegos"]
y
```

Ya tenemos todos nuestros datos listos, ahora tocas seleccionar el modelo que vamos a utilizar (ningun modelo es bueno para todo, se debe seleccionar el modelo depenidendo lo que vayas a hacer)

En este caso vamos a utilizar Arbol de decisiones donde nos dice: El objetivo es crear un modelo que prediga el valor de una variable de destino mediante el aprendizaje de reglas de decisión simples deducidas de las características de los datos.         

Basicamete es predecir un valor por aprender reglas de decision en base a los datos que le pasamos, vamos a intentar predecir el comportamiento si yo le pasara un hombre de 14 años (no esta definido en mi tabla)

```python
# importamos el modulo de tree
from sklearn.tree import DecisionTreeClassifier
# creamos un modelo en base a la clase
modelo = DecisionTreeClassifier()
# cargamos nuestro modelo con los datos de entrada(X) y salida(y)
# fit solo recibe los datos sin nombre de columnas, por lo que hay que modificar X, ya que esa tabla si tiene los nombres
# para eso basta con pedirle que acceda solo a los valores .values
modelo.fit(X.values,y)

# Predicciones
# Lo que espera .predict es una lista de los datos de entrada, en este caso una edad y un genero
# Yo quiero predecir hombres de 14,29 y 30 años
predicciones = modelo.predict([[14,0],[29,0],[30,0]])
predicciones
```

Nos arrojó los siguientes resultados:
- Hombres de 14 años = fps
- Hombres de 29 años = jrpg
- Hombres de 30 años = accion
