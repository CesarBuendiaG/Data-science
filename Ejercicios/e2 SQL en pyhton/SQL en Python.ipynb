```python
import pandas as pd
from sqlalchemy import Engine
from pandasql import sqldf

# Funcion lambda que hace escribir funciones en SQL mucho más facil
pysqldf = lambda q: sqldf(q,globals())
```

```python
df = df = pd.read_csv("abas.csv", encoding='latin-1')
```

```python
df
```

```python
# Lo primero a hacer es renombrar las columnas para quitar los espacios, ya que los espacios rompen el motor de busqueda
df = df.rename(columns={"Códigos de producto": "CodProducto", "Código de cliente": "CodCliente"})
```

```python
# Despues corremos la query con q=""""""

q = """
SELECT
Cliente,
Pedidos,
CodProducto,
Precio,
Total
FROM df
WHERE Pedidos > 300
"""

df_export = pysqldf(q) # Guardamos el resultado en esta variable
```

```python
# Vemos el nuevo dataset filtrado, donde solo estan los pedidos mayores a 300 unidades
df_export
```

```python
# Despues exportamos el archivo en uno nuevo csv
df_export.to_csv("Nuevo_abas.csv")
```
